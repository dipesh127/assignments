```python
# Q2=
import pandas as pd

column=['A','B','C']
index=[1,2,3]
# creating dataframe using dictionary
df = pd.DataFrame([[11,22,33],[44,55,66],[77,88,99]],columns=column,index=index)
changedIndex=list(map(lambda x:x if x==1 else x+2 ,index))
print(df)
print(df.reindex(changedIndex))
```

```python
# Q3=

import pandas as pd
data={'name':["aparna", "pankaj", "sudhir", "Geeku","Ram"],'Values':[10, 20, 30, 40, 50]}
df=pd.DataFrame(data)
valueStore=df['Values']
def sumOfFirstThree():
    sum=0
    count=0
    for i in valueStore:
        count=count+1
        if count<=3:
            sum= sum + i
    print(sum)
    
sumOfFirstThree()
```

```python
# Q4=
import pandas as pd
data={'Text':["hello world","welcome here guys","new year","happy birthday man"]}
df=pd.DataFrame(data)

Word_count=[len(x.split()) for x in df['Text'].tolist()]
df['Word_count']= Word_count
print(df)
```

```python
# Q8=

import pandas as pd
data={'A':[3,8,6,2,9],'B':[5,2,9,3,1],'C':[1,7,4,5,2]}
df=pd.DataFrame(data)
print("Data frame :")
print(df)

print("new Data frame :")
newDf=df[(df['A']>5) & (df['B']<10)]
print(newDf)
```

```python
# Q9=

import pandas as pd
data={'Values':[5, 9, 8, 5, 7, 8, 1, 2, 3,4, 5, 6, 7, 8, 9, 5, 3]}

df=pd.DataFrame(data)
series=df['Values']
print("The mean of the series is :", series.mean())
print("The standard deviation of the series is :", series.std())
print("The median of the series is :",series.median())
```

```python
# Q10=

import pandas as pd
date=['1/1/2023','1/2/2023','1/3/2023','1/4/2023','1/5/2023','1/6/2023','1/7/2023','1/8/2023','1/9/2023','1/10/2023']
sales=[1200,1300,1400,1500,1600,1700,1800,1900,2000,2100]
data=pd.DataFrame({'date':date,'sales':sales})
moving_average=data['sales'].rolling(7).mean()
data['movingAverage']=moving_average
data['current day']=pd.to_datetime('today')
print(data)

```

```python
# Q12=
import pandas as pd
from datetime import datetime

timestamps=[1672531200,1669852800,1667260800,1675209600,1677628800]
df=pd.DataFrame({'timestamps':timestamps})

df['Date']=[datetime.fromtimestamp(x).strftime('%y-%m-%d') for x in df['timestamps']]
print(df)
newdf=df[(df['Date']>='23-01-01') & (df['Date'] <='23-01-31')]
print("\n after computaion:")
print(newdf)
```
